```
# Получите последний рабочий день каждого месяца
import calendar

def last_business_day_of_month(year: int, month: int) -> int:
    return max(calendar.monthcalendar(year, month)[-1:][0][:5])
last_business_days = []
last_business_day_of_month(2020,12)
for i in df_new.index:
  if i.day == last_business_day_of_month(i.year, i.month):
    last_business_days.append(i)
last_business_days
```

### Вычисляем ежедневное процентное закрытие. Первую строчку заполняем нулями
$$ r_t =\dfrac{p_t}{p_{t-1}}-1 $$

```
daily = (df_new['Adj Close'] / df_new['Adj Close'].shift(1) - 1).fillna(0)
daily.head()
```

```
daily.plot()
```

### Вычисляем кумулятивную процентную доходность
$$ i_0=r_0=0 $$
$$ i_t=(i_{t-1}+r_t)100 $$

```
daily_sum = (daily * 100).cumsum()
daily_sum.head()
```

```
daily_sum.plot()
```
